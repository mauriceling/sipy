## Exploratory Data Analysis of a Diabetes Dataset Using SiPy
This case study uses the 403-patient diabetes dataset from

[1] Willems, James P., J. Terry Saunders, Dawn E. Hunt, and John B. Schorling. "Prevalence of coronary heart disease risk factors among rural blacks: a community-based study." Southern medical journal 90, no. 8 (1997): 814-820.

[2] Schorling, John B., Julienne Roach, Marjorie Siegel, Natalie Baturka, Dawn E. Hunt, Thomas M. Guterbock, and Herbert L. Stewart. "A trial of church-based smoking cessation interventions for rural African Americans."Preventive Medicine 26, no. 1 (1997): 92-101. 

```bash
# read the data
read csv db from data/diabetes.csv
```

```bash
# generate a heatmap of correlations from the data
plot heatmap dataframe data=db
```

Heatmap suggests the following which may be interesting
1. High correlation between waist and hip measurement, which is expected
2. HDL has low correlation to waist and hip measurement, which is less expected

So we are going to 
1. Look at the distributions of waist, hip, and HDL measurements
2. Run correlation tests and plot scatterplots between these 3 attributes

```bash
plot histogram dataframe data=db.waist bins=10 kde=true
plot histogram dataframe data=db.hip bins=10 kde=true
plot histogram dataframe data=db.hdl bins=10 kde=true
```

```bash
# do a correlation test between waist and hip measurements, and generate a scatterplot
correlate pearson dataframe wide data=db.waist,db.hip
plot scatterplot dataframe data=db x=waist y=hip
```

```bash
# do a correlation test between waist and hdl measurements, and generate a scatterplot
correlate pearson dataframe wide data=db.waist,db.hdl
plot scatterplot dataframe data=db x=waist y=hdl
```

```bash
# do a correlation test between hip and hdl measurements, and generate a scatterplot
correlate pearson dataframe wide data=db.hip,db.hdl
plot scatterplot dataframe data=db x=hip y=hdl
```

Correlation analysis shows
1. Highly significant correlation between waist and hip measurements (r = 0.83708; 95% CI = [0.81, 0.86]; p-value = 1.45e-106)
2. Lesser and inverse but significant correlations between waist and HDL measurements (r = -0.268369; 95% CI = [-0.36, -0.17]; p-value = 5.01e-08), and hip and HDL measurements (r = -0.21006; 95% CI = [-0.3, -0.11]; p-value = 0.000023)
